```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("="*50)
print("ZOMATO ANALYSIS (Using Sample Data)")
print("="*50)

# Creating sample restaurant data
np.random.seed(42)
data = {
    'restaurant_name': [f'Restaurant_{i}' for i in range(100)],
    'cuisines': np.random.choice(['North Indian', 'Chinese', 'Italian', 'South Indian', 'Fast Food', 'Cafe', 'Biryani'], 100),
    'rating': np.random.uniform(2.5, 4.9, 100),
    'cost': np.random.randint(200, 3000, 100),
    'location': np.random.choice(['Connaught Place', 'Rajouri Garden', 'Karol Bagh', 'South Delhi', 'Chanakyapuri', 'Nehru Place', 'Lajpat Nagar'], 100)
}
df = pd.DataFrame(data)

# Round ratings
df['rating'] = df['rating'].round(1)

print(f"📊 Total restaurants: {len(df)}")
print(f"⭐ Average rating: {df['rating'].mean():.2f}")
print(f"💰 Average cost: ₹{df['cost'].mean():.0f}")

# PICTURE 1: Rating Distribution
plt.figure(figsize=(10,5))
plt.hist(df['rating'], bins=15, color='orange', edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.title('📊 Restaurant Ratings Distribution')
plt.show()

# PICTURE 2: Top Cuisines
top_cuisines = df['cuisines'].value_counts().head(5)
plt.figure(figsize=(10,5))
top_cuisines.plot(kind='bar', color='green', edgecolor='black')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('🍜 Top 5 Cuisines')
plt.xticks(rotation=45)
plt.show()

print("\n🏆 TOP 5 CUISINES:")
print(top_cuisines)

# PICTURE 3: Price vs Rating
plt.figure(figsize=(10,5))
plt.scatter(df['cost'], df['rating'], alpha=0.6, color='red')
plt.xlabel('Cost for Two (₹)')
plt.ylabel('Rating')
plt.title('💰 Price vs Rating')
plt.show()

# PICTURE 4: Top Locations
top_locations = df['location'].value_counts().head(10)
plt.figure(figsize=(10,5))
top_locations.plot(kind='barh', color='purple', edgecolor='black')
plt.xlabel('Number of Restaurants')
plt.title('📍 Top Locations')
plt.show()

print("\n📍 TOP LOCATIONS:")
print(top_locations.head(3))

# Recommendations
print("\n" + "="*50)
print("💡 5 RECOMMENDATIONS FOR ALFIDO TECH")
print("="*50)
print(f"1. Partner with {top_cuisines.index[0]} restaurants")
print(f"2. Focus on {top_locations.index[0]} area")
print("3. Keep prices between ₹500-₹1500")
print("4. Add 'Top Rated Near Me' feature")
print("5. Create cuisine comparison guides")

print("\n" + "="*50)
print("✅ ANALYSIS COMPLETE!")
print("📁 Click File → Download → Download .ipynb to save")
print("="*50)
```
